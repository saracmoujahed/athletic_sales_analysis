```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_20 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_21 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_20.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_21.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_20.info()
```

```python
# Check the 2021 sales data types.
athletic_sales_21.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_joined = pd.concat([athletic_sales_20, athletic_sales_21], ignore_index=True)
athletic_sales_joined
```

```python
# Check if any values are null.
athletic_sales_joined.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_joined.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_joined = athletic_sales_joined.astype({"invoice_date": "datetime64[ns]"})
athletic_sales_joined
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_joined.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
athletic_sales_grouped = athletic_sales_joined.groupby(["region", "state", "city"]).agg({"units_sold": "sum"})
athletic_sales_grouped = athletic_sales_grouped.rename(columns={"units_sold" : "Total_Products_Sold"})
# Show the top 5 results.
athletic_sales_grouped_top_5 = athletic_sales_grouped.sort_values(by="Total_Products_Sold", ascending=False)
athletic_sales_grouped_top_5.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
athletic_sales_pivot = athletic_sales_joined.pivot_table(
    values="units_sold", index=["region", "state", "city"], aggfunc="sum")

# Rename the "units_sold" column to "Total_Products_Sold"
athletic_sales_pivot = athletic_sales_pivot.rename(columns={"units_sold": "Total_Products_Sold"})
# Show the top 5 results.
athletic_sales_grouped_top_5 = athletic_sales_pivot.sort_values(by="Total_Products_Sold", ascending=False)
athletic_sales_grouped_top_5.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_grouped = athletic_sales_joined.groupby(["region", "state", "city"]).agg({"total_sales": "sum"})
athletic_sales_grouped = athletic_sales_grouped.sort_values(by="total_sales", ascending=False)
# Rename the "total_sales" column to "Total Sales"
athletic_sales_grouped_top_5 = athletic_sales_grouped.rename(columns={"total_sales" : "Total Sales"})
# Show the top 5 results.
athletic_sales_grouped_top_5.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_pivot = athletic_sales_joined.pivot_table(
    values="total_sales", index=["region", "state", "city"], aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_pivot = athletic_sales_pivot.rename(columns={"total_sales": "Total Sales"})


# Show the top 5 results.
athletic_sales_pivot_top_5 = athletic_sales_pivot.sort_values(by="Total Sales", ascending=False)
athletic_sales_pivot_top_5.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
athletic_sales_grouped_retailer = athletic_sales_joined.groupby(["retailer", "region", "state", "city"]).agg({"total_sales": "sum"})

# Rename the "total_sales" column to "Total Sales"
athletic_sales_grouped_retailer = athletic_sales_grouped_retailer.rename(columns={"total_sales": "Total Sales"})
# Show the top 5 results.
athletic_sales_grouped_retailer = athletic_sales_grouped_retailer.sort_values(by="Total Sales", ascending=False)
athletic_sales_grouped_retailer.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
athletic_sales_pivot_retailer = athletic_sales_joined.pivot_table(
    values="total_sales", index=["retailer", "region", "state", "city"], aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_pivot_retailer = athletic_sales_pivot_retailer.rename(columns={"total_sales": "Total Sales"})

# Show the top 5 results.
athletic_sales_pivot_retailer = athletic_sales_pivot_retailer.sort_values(by="Total Sales", ascending=False)
athletic_sales_pivot_retailer.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
athletic_womens_sales_retailer = athletic_womens_sales.groupby(["retailer", "region", "state", "city"]).agg({"units_sold": "sum"})

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
athletic_womens_sales_retailer = athletic_womens_sales_retailer.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})


# Show the top 5 results.
athletic_womens_sales_retailer_5 = athletic_womens_sales_retailer.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)
athletic_womens_sales_retailer_5.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
athletic_sales_invoice = athletic_sales_joined.pivot_table(
    values="total_sales", index=["invoice_date"], aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_invoice = athletic_sales_invoice.rename(columns={"total_sales": "Total Sales"})


# Show the table.
athletic_sales_invoice.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
athletic_sales_invoice.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
athletic_sales_invoice.sort_values(by="Total Sales", ascending=False)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
athletic_sales_invoice.resample('W').sum()


# Sort the resampled pivot table in descending order on "Total Sales".
athletic_sales_invoice.sort_values(by="Total Sales", ascending=False)
```
